# Results

## Hydrogen molecule dataset

Following [1], we begin our study by constructing a Hohenberg-Kohn map for the hydrogen molecule $\text{H}_2$. The only geometrical degree of freedom for this molecule is the separation $R$ between the H atoms. In order to check our calculations against Ref. [1], we generated a similar dataset, choosing $150$ uniformly spaced values of $R$ ranging between $0.5$ and $1.5 \text{ Å}$. Figure 1 plots the total energy of the molecule as a function of $R$, showing a computed equilibrium separation of $\approx0.75 \text{ Å}$ in reasonable agreement with the experimental value of $0.7414 \text{ Å}$ [2]. The calculated energies are compared against the values in [1], which were computed with Quantum ESPRESSO using the PBE XC functional and PAWs, with a Troullier–Martins pseudopotential for describing the ionic cores. The energies from [1] were multiplied by an overall scaling factor in order to compare the calculations. We note that there is slight disagreement in the energies and in the location of the equilibrium separation due to differences in the pseudopotential.

```python
# read in calculation data (computed elsewhere)
str_pref = os.environ['PROJDIR']+'/data/H2_DFT/temp_data/store/'
test = np.load(os.environ['PROJDIR']+'/data/H2_DFT/temp_data/store/sep_store/sep149.npy')

# read in the calculation data
sim_no = 150
ens = []
seps= []
fours = []

min_en = 0
eq_sep = 0

for n in range(sim_no):
    # load separation, energy, and density
    sep = np.load(str_pref+'sep_store/sep'+str(n)+'.npy')
    en = np.load(str_pref+'en_store/en'+str(n)+'.npy')
    four = np.load(str_pref+'four_store/four'+str(n)+'.npy')

    # put results in a nicer format
    sep = np.reshape(sep,(1,))[0]
    en = np.reshape(en,(1,))[0]['energy']
    four = np.real(four)
    
    # get min energy and equilibrium separation
    if en < min_en:
        min_en = en
        eq_sep = sep

    # store quantities
    ens.append(en)
    seps.append(sep)
    fours.append(four)

print('The minimum energy is '+str(min_en)+' eV. The equilibrium separation is '+str(eq_sep)+' Angstrom.')
```

```python
# return to default and make the fontbigger
plt.rcParams.update(plt.rcParamsDefault)
font = {'size'   : 15}
plt.rc('font', **font)

# create figure
figure = plt.figure()
ax = figure.add_subplot(1,1,1)

# add computed energies
ax.plot(seps, ens, label='computed energies')

# add energies from online dataset
fac = 0.04335 # apply appropriate scale factor
#fac = 0.0434
ax.plot(rep_seps, rep_ens * fac, 'kx', MarkerSize = 10, label = 'energies from [1]')

ax.set_xlabel('separation (Å)')
ax.set_ylabel('energy (eV)')
ax.legend()
plt.show()
```

Figure 1: The total energy of the H$_2$ molecule versus the separation between the H atoms. The energies of $150$ molecular configurations corresponding to uniformly spaced separations between $0.5$ and $1.5 \text{ Å}$ are plotted in blue. ## KRR model of the electron density

For each of the $150$ molecular configurations in the dataset, the real-space ground state electron density was computed on a $100\times100\times100$ uniform grid within a box of length  $20 \text{ bohr}$. The density was then Fourier transformed, and a $25\times25\times25$ grid of Fourier coefficients was used to represent the density in the Hohenberg-Kohn mapping. A separate KRR model was trained for each Fourier component with hyperparameters set equal to $\sigma = 1$ and $\gamma = 0$. We note that these hyperparmeters may be optimized in each model through cross-validation, a procedure that we plan to implement in the Final Report.

In order to evaluate the performance of the models, the dataset was split into a training set and a test set, each containing $75$ molecular configurations with separations uniformly spaced from $0.5 \text{ Å}$ to $1.5 \text{ Å}$.

```python
# split dataset into training set and test set
train_set = np.array(seps[0:150:2])
test_set = np.array(seps[1:150:2])

train_ens = np.array(ens[0:150:2])
test_ens = np.array(ens[1:150:2])

train_fours = fours[0:150:2]
test_fours = fours[1:150:2]
```

```python
# get master kernel (takes awhile, but speeds things up downstream)
sim_no = 150
    
# create list of gaussian potentials
mast_pots = []
grid_len = 5.29177*2

for n in range(sim_no):
    dist = seps[n]
    pot = pot_rep(dist, grid_len)
    mast_pots.append(pot)
    
# get kernel matrix (computationally expensive!)
sig = 1
mast_kern = get_kern_mat(mast_pots, sig)
```

```python
# get test kernel
sim_no = 75
    
# create list of gaussian potentials
pots = []
grid_len = 5.29177*2

for n in range(sim_no):
    dist = train_set[n]
    pot = pot_rep(dist, grid_len)
    pots.append(pot)
    
# get kernel matrix (computationally expensive!)
sig = 1
kern = get_kern_mat(pots, sig)
```

```python
# as a test, first model a single fourier component

# specify fourier component to model
four1 = 0
four2 = 1
four3 = 0
data_vec = []

# get chosen Fourier component for each training density 
for n in range(sim_no):
    comp = train_fours[n][four1, four2, four3]
    data_vec.append(comp)

# plot the Fourier component
plt.plot(data_vec)
plt.show()

# build the model
test_ind = 12
lam = 0
test = get_beta(kern, data_vec, lam=lam)

# test the model out 
pot_test = pots[test_ind]
pred_test=pred_four(pot_test,pots,test,sig=1)
print(pred_test)
print(train_fours[test_ind][four1, four2, four3])
```

```python
# for a given number of training points, construct HK and KS maps
M = 7 # training points

# define training and test indices
train_indices = [int(n) for n in np.round(np.linspace(0,149,M))]
test_indices = [n for n in range(150) if n not in train_indices]

# define train and test sets
train_set = np.array([seps[n] for n in train_indices])
test_set = np.array([seps[n] for n in test_indices])

train_ens = np.array([ens[n] for n in train_indices])
test_ens = np.array([ens[n] for n in test_indices])

train_fours = np.array([fours[n] for n in train_indices])
test_fours = np.array([fours[n] for n in test_indices])

# get potential kernel
pots = []
grid_len = 5.29177*2
for n in range(M):
    dist = train_set[n]
    pot = pot_rep(dist, grid_len)
    pots.append(pot)  
    
# get kernel matrix 
sig = 30
kern = get_kern_mat(pots, sig)

# create test potentials
test_pots = []
for n in range(len(test_indices)):
    dist = test_set[n]
    pot = pot_rep(dist, grid_len)
    test_pots.append(pot)

# build models for each Fourier component
betas = []
no_four = 25
lam = 0
for i in range(no_four):
    four1 = i
    for j in range(no_four):
        four2 = j
        for k in range(no_four):
            four3 = k

            data_vec = []

            # get chosen Fourier component for each training density 
            for n in range(M):
                comp = train_fours[n][four1, four2, four3]
                data_vec.append(comp)

            # build the model
            beta = get_beta(kern, data_vec, lam=lam)
            betas.append(beta)
            
# get kernel matrix for densities
sig_en = 2000
dens_kern = get_kern_mat(train_fours, sig_en)

# train density to energy model
lam = 0
beta_en = get_beta(dens_kern, train_ens, lam=lam)

# train potential to energy model
lam =0
beta_KS = get_beta(kern, train_ens, lam=lam)
```

```python
# compare errors
HK_errs = []
KS_errs = []

for n in range(len(test_pots)):
    # choose test potential
    test_ind = n
    pot_test = test_pots[n]

    # get distances from training points
    dists = get_dist(pot_test,pots,sig=sig)

    # get density
    no_four = 25
    test_dens = np.zeros([no_four, no_four, no_four])
    count = 0
    for i in range(no_four):
        for j in range(no_four):
            for k in range(no_four):
                # use model to calculate density
                test_dens[i,j,k]=pred_fast(dists,betas[count])

                count+=1

    # get energy
    true_en = test_ens[test_ind]
    test_en = pred_four(test_dens,train_fours,beta_en,sig=sig_en)
    HK_err = np.abs(true_en-test_en)
    HK_errs.append(HK_err)


    # test KS map
    pred_test=pred_four(pot_test,pots,beta_KS,sig=sig)
    KS_err = np.abs(true_en - pred_test)
    KS_errs.append(KS_err)
    
```

```python
fig, ax = plt.subplots(1,1)
ax.plot(test_set, KS_errs,label='KS')
ax.plot(test_set, HK_errs,label='HK')
ax.set_xlabel('separation (Å)')
ax.set_ylabel('log error')
ax.set_yscale('log')
plt.legend()
plt.show()
```

```python
# play around with cross validation
sigs = [100]
# store_preds = []
grid_len = 5.29177*2
lam = 0
for n in range(len(sigs)):
    sig_curr = sigs[n]
    kern = get_kern_mat(pots, sig_curr)
    beta_KS = get_beta(kern, train_ens, lam=lam)
    store = []
    for m in range(len(seps)):
        sep = seps[m]
        pot = pot_rep(sep, grid_len)
        pred_test=pred_four(pot,pots,beta_KS,sig=sig_curr)
        store.append(pred_test)
    store_preds.append(store)

    print(n)
```
