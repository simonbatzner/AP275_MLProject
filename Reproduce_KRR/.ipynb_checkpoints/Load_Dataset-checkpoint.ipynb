```python
import matplotlib.pyplot as plt
import os
import numpy as np
import time
import scipy as sp
from scipy.fftpack import *
```

## Load dataset.

```python
# load MD results
str_pref = os.environ['WORKDIR']+'/Al_Data/Store/'
MD_pos = np.load(str_pref+'MD_positions.npy')
MD_ens = np.load(str_pref+'MD_energies.npy')
```

```python
# load DFT results
str_pref = os.environ['WORKDIR']+'/Al_Data/Store/'
no_sim = 201

pos = []
ens = []
fours = []

for n in range(no_sim):
    # load arrays
    en_curr = np.reshape(np.load(str_pref+'en_store/energy'+str(n)+'.npy'),(1))[0]
    pos_curr = np.load(str_pref+'pos_store/pos'+str(n)+'.npy')
    four_curr = np.load(str_pref+'four_store/four'+str(n)+'.npy')
    
    # store arrays
    ens.append(en_curr)
    pos.append(pos_curr)
    fours.append(four_curr)

# convert to np arrays
ens = np.array(ens)
pos = np.array(pos)
fours = np.array(fours)
```

```python
plt.figure()
plt.plot(ens)
plt.show()
```
