```python
import gptools
import numpy as np
import matplotlib.pyplot as plt
```

```python
import warnings
warnings.filterwarnings("ignore")
```

Note: small modifications were made to the gptools library to make it compatible with Python 3.

### Make 1D dataset.

```python
# make dataset
dl = 0
dr = 2*np.pi
xs = np.linspace(dl,dr,100)
ys = np.cos(xs)
dervs = -np.sin(xs)

# make training set
tl = np.pi/4
tr = 7*np.pi/4
tp_no = 5
tp = np.linspace(tl, tr, tp_no)
tp_vals = np.cos(tp)
```

```python
# plot dataset
plt.plot(xs, ys, label='function')
plt.plot(xs, dervs, label='derivative')
plt.plot(tp, tp_vals, 'kx', label='training points')
plt.legend()
plt.show()
```

### Make 1D GP model with no derivative observations.

The following code is based on [this](https://github.com/markchil/gptools/blob/master/demo/demo.py) tutorial.

```python
# choose covariance kernel
k_SE = gptools.SquaredExponentialKernel(param_bounds=[(0, 20), (0, 5)])

# make gp model
gp = gptools.GaussianProcess(k_SE)

# add training data
gp.add_data(tp, tp_vals)

# optimize hyperparameters with MAP estimate
gp.optimize_hyperparameters(random_starts=10)

# print hyperparameters (prefactor, lengths scale, noise)
print(gp.params[:])
```

```python
# make predictions with GP model
x_star = np.linspace(dl, dr, 100)
y_star, err_y_star = gp.predict(x_star)
```

```python
# plot predictions and estimated error
mag = 2 # size of uncertainty envelope
plt.plot(x_star, y_star, label='predicted')
plt.plot(xs, ys, label='true')
plt.plot(tp, tp_vals, 'kx', label='training set')


plt.fill(np.concatenate([x_star, x_star[::-1]]),
         np.concatenate([y_star - mag * err_y_star,
                         (y_star + mag * err_y_star)[::-1]]),
         alpha=.2, fc='b', ec='None', label='uncertainty')

plt.legend()
plt.show()
```

```python
# make derivative predictions with GP model
x_star = np.linspace(dl, dr, 100)
y_derv_star, err_y_derv_star = gp.predict(x_star, n=1)
```

```python
# plot predictions and estimated error
mag = 2 # size of uncertainty envelope
plt.plot(x_star, y_derv_star, label='predicted')
plt.plot(xs, dervs, label='true')
plt.plot(tp, -np.sin(tp), 'kx', label='training set')


plt.fill(np.concatenate([x_star, x_star[::-1]]),
         np.concatenate([y_derv_star - mag * err_y_derv_star,
                         (y_derv_star + mag * err_y_derv_star)[::-1]]),
         alpha=.2, fc='b', ec='None', label='uncertainty')

plt.legend()
plt.show()
```

### Make 1D GP model with derivative observations.

```python
# get derivative between two training points
x_derv = (tp[2] + tp[3])/2
y_derv = -np.sin(x_derv)

# make gp model
gp2 = gptools.GaussianProcess(k_SE)

# add training data
gp2.add_data(tp, tp_vals)

# add derivative to training set
gp2.add_data(x_derv, y_derv, n=1)

# optimize hyperparameters with MAP estimate
gp2.optimize_hyperparameters(random_starts=10)

# print hyperparameters (prefactor, lengths scale, noise)
print(gp2.params[:])
```

```python
# make predictions with GP model
x_star = np.linspace(dl, dr, 100)
y_star, err_y_star = gp2.predict(x_star)
```

```python
# plot predictions and estimated error
mag = 10 # size of uncertainty envelope
plt.plot(x_star, y_star, label='predicted')
plt.plot(xs, ys, label='true')
plt.plot(tp, tp_vals, 'kx', label='training set')
plt.plot(x_derv, np.cos(x_derv), 'k+', label='derivative pt')


plt.fill(np.concatenate([x_star, x_star[::-1]]),
         np.concatenate([y_star - mag * err_y_star,
                         (y_star + mag * err_y_star)[::-1]]),
         alpha=.2, fc='b', ec='None', label='uncertainty')

plt.legend()
plt.show()
```

```python
# make derivative predictions with GP model
x_star = np.linspace(dl, dr, 100)
y_derv_star, err_y_derv_star = gp2.predict(x_star, n=1)
```

```python
# plot predictions and estimated error
mag = 2 # size of uncertainty envelope
plt.plot(x_star, y_derv_star, label='predicted')
plt.plot(xs, dervs, label='true')
plt.plot(tp, -np.sin(tp), 'kx', label='training set')


plt.fill(np.concatenate([x_star, x_star[::-1]]),
         np.concatenate([y_derv_star - mag * err_y_derv_star,
                         (y_derv_star + mag * err_y_derv_star)[::-1]]),
         alpha=.2, fc='b', ec='None', label='uncertainty')

plt.legend()
plt.show()
```

### Make 2D GP model with no derivative observations.

```python
def gauss(vec):
    val = np.exp(-(vec[0]**2/2 + vec[1]**2/2))
    return val

def gauss_func(x, y):
    val = np.exp(-(x**2/2 + y**2/2))
    return val

def gauss_grad(vec):
    x_derv = -vec[0] * np.exp(-(vec[0]**2/2 + vec[1]**2/2))
    y_derv = -vec[1] * np.exp(-(vec[0]**2/2 + vec[1]**2/2))
    return np.array([x_derv, y_derv])

def derv_xy(vec):
    val = vec[0]*vec[1] * np.exp(-(vec[0]**2/2 + vec[1]**2/2))
    return val

def derv_x(vec):
    val = -vec[0] * np.exp(-(vec[0]**2/2 + vec[1]**2/2))
    return val

def derv_y(vec):
    val = -vec[1] * np.exp(-(vec[0]**2/2 + vec[1]**2/2))
    return val

def derv_x_func(x, y):
    val = -x * np.exp(-(x**2/2 + y**2/2))
    return val

def derv_y_func(x, y):
    val = -y * np.exp(-(x**2/2 + y**2/2))
    return val
```

```python
# plot contour
fig, ax = plt.subplots(1,1)
xs = np.linspace(-2, 2, 100)
ys = np.linspace(-2, 2, 101)
xs_mesh, ys_mesh = np.meshgrid(xs, ys)
zs = gauss_func(xs_mesh, ys_mesh)

CS = ax.contourf(xs_mesh, ys_mesh, zs,100)
plt.colorbar(CS)
plt.title('function to be learned')
plt.show()
```

```python
# plot y derivatives
fig, ax = plt.subplots(1,1)
xs = np.linspace(-2, 2, 100)
ys = np.linspace(-2, 2, 101)
xs_mesh, ys_mesh = np.meshgrid(xs, ys)
dervs_y = derv_y_func(xs_mesh, ys_mesh)

CS = ax.contourf(xs_mesh, ys_mesh, dervs_y,100)
plt.colorbar(CS)
plt.title('function to be learned')
plt.show()
```

```python
# get training points
vals = np.array([-2, -1, 0, 1, 2])
vecs = []
fvals = []
for m in range(len(vals)):
    x_val = vals[m]  
    for n in range(len(vals)):
        y_val = vals[n]
        
        ax.plot(x_val, y_val, 'kx')
        
        vec_curr = np.array([x_val, y_val])
        vecs.append(vec_curr)
        fvals.append(gauss(vec_curr))
        
vecs = np.array(vecs)
fvals = np.array(fvals)
```

```python
# choose covariance kernel
k_SE = gptools.SquaredExponentialKernel(param_bounds=[(0, 20), (0, 5), (0, 5)],\
                                        num_dim = 2)

# make gp model
gp3 = gptools.GaussianProcess(k_SE)

# add training data
gp3.add_data(vecs, fvals)

# optimize hyperparameters with MAP estimate
gp3.optimize_hyperparameters(random_starts=20)

# print hyperparameters (prefactor, lengths scale, noise)
print(gp3.params[:])
```

```python
# predict values
pred = np.zeros(xs_mesh.shape)
err = np.zeros(xs_mesh.shape)
yderv_pred = np.zeros(xs_mesh.shape)
yerr = np.zeros(xs_mesh.shape)
for m in range(xs_mesh.shape[0]):
    for n in range(xs_mesh.shape[1]):
        vec_curr = np.array([xs_mesh[m,n], ys_mesh[m,n]])
        
        # predict function value and y derivative
        pred_full = gp3.predict(vec_curr, [0,0])
        ypred_full = gp3.predict(vec_curr, [0,1])
        
        # record values
        pred_curr = pred_full[0][0]
        err_curr = pred_full[1][0]
        ypred_curr = ypred_full[0][0]
        yerr_curr = ypred_full[1][0]
        
        pred[m,n]=pred_curr
        err[m,n]=err_curr
        yderv_pred[m,n]=ypred_curr
        yerr[m,n]=yerr_curr
```

```python
# plot GP predictions
fig, ax = plt.subplots(1,1)

CS = ax.contourf(xs_mesh, ys_mesh, pred,100)
plt.colorbar(CS)
plt.title('GP predictions')
plt.show()
```

```python
# plot GP error
fig, ax = plt.subplots(1,1)

CS = ax.contourf(xs_mesh, ys_mesh, err,100)
plt.colorbar(CS)
plt.title('predicted error')
plt.show()
```

```python
# plot true error
fig, ax = plt.subplots(1,1)
CS = ax.contourf(xs_mesh, ys_mesh, np.abs(zs-pred),100)
plt.colorbar(CS)
plt.title('true error')
plt.show()
```

```python
# plot y derv predictions
fig, ax = plt.subplots(1,1)

CS = ax.contourf(xs_mesh, ys_mesh, yderv_pred,100)
plt.colorbar(CS)
plt.title('GP predictions')
plt.show()
```

```python
# plot y derv predicted error
fig, ax = plt.subplots(1,1)

CS = ax.contourf(xs_mesh, ys_mesh, yerr,100)
plt.colorbar(CS)
plt.title('GP predictions')
plt.show()
```

```python
# plot true derivative error
fig, ax = plt.subplots(1,1)
CS = ax.contourf(xs_mesh, ys_mesh, np.abs(dervs_y-yderv_pred),100)
plt.colorbar(CS)
plt.title('true error')
plt.show()
```

### Make 2D GP model with derivative observations.

```python
# choose covariance kernel
k_SE = gptools.SquaredExponentialKernel(param_bounds=[(0, 20), (0, 5), (0, 5)],\
                                        num_dim = 2)

# make gp model
gp4 = gptools.GaussianProcess(k_SE)

# add training data
gp4.add_data(vecs, fvals)

# add derivative information
derv_vec = np.array([0,-1.25])
vec_grad = gauss_grad(derv_vec)
gp4.add_data(derv_vec, vec_grad[1],n=[0,1])

# optimize hyperparameters with MAP estimate
gp4.optimize_hyperparameters(random_starts=20)

# print hyperparameters (prefactor, lengths scale, noise)
print(gp4.params[:])
```

```python
# predict values
pred = np.zeros(xs_mesh.shape)
err = np.zeros(xs_mesh.shape)
yderv_pred = np.zeros(xs_mesh.shape)
yerr = np.zeros(xs_mesh.shape)
for m in range(xs_mesh.shape[0]):
    for n in range(xs_mesh.shape[1]):
        vec_curr = np.array([xs_mesh[m,n], ys_mesh[m,n]])
        
        # predict function value and y derivative
        pred_full = gp4.predict(vec_curr, [0,0])
        ypred_full = gp4.predict(vec_curr, [0,1])
        
        # record values
        pred_curr = pred_full[0][0]
        err_curr = pred_full[1][0]
        ypred_curr = ypred_full[0][0]
        yerr_curr = ypred_full[1][0]
        
        pred[m,n]=pred_curr
        err[m,n]=err_curr
        yderv_pred[m,n]=ypred_curr
        yerr[m,n]=yerr_curr
```

```python
# plot GP error
fig, ax = plt.subplots(1,1)

CS = ax.contourf(xs_mesh, ys_mesh, err,100)
plt.colorbar(CS)
plt.title('predicted error')
plt.show()
```

```python
# plot true error
fig, ax = plt.subplots(1,1)
CS = ax.contourf(xs_mesh, ys_mesh, np.abs(zs-pred),100)
plt.colorbar(CS)
plt.title('true error')
plt.show()
```

```python
# plot y derv predicted error
fig, ax = plt.subplots(1,1)

CS = ax.contourf(xs_mesh, ys_mesh, yerr,100)
plt.colorbar(CS)
plt.title('GP predictions')
plt.show()
```

```python
# plot true derivative error
fig, ax = plt.subplots(1,1)
CS = ax.contourf(xs_mesh, ys_mesh, np.abs(dervs_y-yderv_pred),100)
plt.colorbar(CS)
plt.title('true error')
plt.show()
```
