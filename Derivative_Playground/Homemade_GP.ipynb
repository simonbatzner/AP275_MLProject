### Make 1D dataset.

```python
# make dataset
dl = 0
dr = 2*np.pi
xs = np.linspace(dl,dr,100)
ys = np.cos(xs)
dervs = -np.sin(xs)

# make training set
tl = np.pi/4
tr = 7*np.pi/4
tp_no = 5
tp = np.linspace(tl, tr, tp_no)
tp_vals = np.cos(tp)
```

```python
# plot dataset
plt.plot(xs, ys, label='function')
plt.plot(xs, dervs, label='derivative')
plt.plot(tp, tp_vals, 'kx', label='training points')
plt.legend()
plt.show()
```

### Make GP model with gptools.

```python
# choose covariance kernel
k_SE = gptools.SquaredExponentialKernel(param_bounds=[(0, 20), (0, 5)])

# make gp model
gp = gptools.GaussianProcess(k_SE)

# add training data
gp.add_data(tp, tp_vals)

# optimize hyperparameters with MAP estimate
gp.optimize_hyperparameters(random_starts=10)

# print hyperparameters (prefactor, lengths scale, noise)
print(gp.params[:])
```

```python
# make predictions with GP model
x_star = np.linspace(dl, dr, 100)
y_star, err_y_star = gp.predict(x_star)

# plot predictions and estimated error
mag = 2 # size of uncertainty envelope
plt.plot(x_star, y_star, label='predicted')
plt.plot(xs, ys, label='true')
plt.plot(tp, tp_vals, 'kx', label='training set')


plt.fill(np.concatenate([x_star, x_star[::-1]]),
         np.concatenate([y_star - mag * err_y_star,
                         (y_star + mag * err_y_star)[::-1]]),
         alpha=.2, fc='b', ec='None', label='uncertainty')

plt.legend()
plt.show()
```

### Create GP model from scratch.

The code in this section follows Algorithm 2.1 (pg. 19) of "Gaussian Processes for Machine Learning" by Rasmussen and Williams.

```python
# get squared exponential covariance between two input vectors
def SE_cov(x1, x2, sig, ls):
    k = sig**2*np.exp(-np.sum((x1-x2)**2)/(2*ls**2))
    return k
```

```python
# get nxn noiseless kernel matrix
# X is Dxn design matrix
def get_SE_K(X,sig,ls):
    nod = X.shape[0] # number of dimensions
    noi = X.shape[1] # number of inputs
    
    # create diagonal
    K = np.diag(sig**2*np.ones(noi))
    
    # calculate off diagonals
    for m in range(noi):
        for n in range(m+1, noi):
            cov = SE_cov(X[:,m],X[:,n],sig,ls)
            K[m,n] = cov
            K[n,m] = cov
    
    # perform cholesky decomposition
    L = np.linalg.cholesky(K)
    
    return K, L
```

```python
# get kernel vector
def get_SE_kv(X,x,sig,ls):
    kv = np.zeros([X.shape[1],1])
    for m in range(X.shape[1]):
        kv[m] = SE_cov(X[:,m],x,sig,ls)
        
    return kv
```

```python
# get alpha
def GP_SE_alpha(K,L,y): 
    # get alpha
    ts1 = sp.linalg.solve_triangular(L,y,lower=True)
    alpha = sp.linalg.solve_triangular(L.transpose(),ts1)
            
    return alpha
```

```python
# get likelihood
def GP_SE_like(K,L,y,alpha): 
    # get log marginal likelihood
    like = -(1/2)*np.matmul(y.transpose(),alpha)-\
            np.sum(np.log(np.diagonal(L)))-\
            np.log(2*np.pi)*K.shape[1]/2
            
    return like

# get likelihood as a function of hyperparameters
def like_hyp(hyp,X,y):
    # unpack hyperparameters
    sig = hyp[0]
    ls = hyp[1]
    
    # calculate likelihood
    K, L = get_SE_K(X,sig,ls)
    alpha = GP_SE_alpha(K,L,y)
    like = GP_SE_like(K,L,y,alpha)
    
    return like

# get minus likelihood as a function of hyperparameters
def minus_like_hyp(hyp,X,y):
    like = like_hyp(hyp,X,y)
    minus_like = -like
    return minus_like
```

```python
# make GP prediction with SE kernel
def GP_SE_pred(X,y,K,L,alpha,sig,ls,xt):
    # get kernel vector
    kv = get_SE_kv(X,xt,sig,ls)
    
    # get predictive mean
    f = np.matmul(kv.transpose(),alpha)
    
    # get predictive variance
    v = sp.linalg.solve_triangular(L,kv,lower=True)
    var = sig**2 - np.matmul(v.transpose(),v)
    
    return f, var
```

### Try homemade GP on a simple example.

```python
# use hyperparameters from gptools
sig = 1.1285086112439977
ls = 2.0211594230721492
```

```python
# make dataset
dl = 0
dr = 2*np.pi
xs = np.linspace(dl,dr,100)
ys = np.cos(xs)
dervs = -np.sin(xs)

# make training set
tl = np.pi/4
tr = 7*np.pi/4
tp_no = 5
tp = np.linspace(tl, tr, tp_no).reshape(1,tp_no)
tp_vals = np.cos(tp).reshape(tp_no,1)

# get K and L
K, L = get_SE_K(tp,sig,ls)

# get alpha and likelihood
alpha = GP_SE_alpha(K,L,tp_vals)
like = GP_SE_like(K,L,tp_vals,alpha)
```

```python
# make prediction
xt = 0
f, var = GP_SE_pred(tp,tp_vals,K,L,alpha,sig,ls,xt)

print('homemade prediction:')
print('prediction: '+str(f[0][0]))
print('variance: '+str(var[0][0]))

# # compare with gptools model
# y_star, err_y_star = gp.predict(xt)

# print('\ngptools prediction:')
# print('prediction: '+str(y_star[0]))
# print('variance: '+str(err_y_star[0]**2))
```

### Plot marginal likelihood.

```python
sigs = np.linspace(0.95,1.3,100)
lens = np.linspace(1.5,2.5,100)
sig_mesh, len_mesh = np.meshgrid(sigs, lens)
like_cont = np.zeros(sig_mesh.shape)

max_like = -20

for m in range(len(lens)):
    for n in range(len(sigs)):
        sig_cur = sig_mesh[m][n]
        len_cur = len_mesh[m][n]
        hyp = np.array([sig_cur, len_cur])
        
        # get likelihood
        like = like_hyp(hyp,tp,tp_vals)
        like_cont[m][n] = like[0][0]
        
        if like > max_like:
            best_sig = sig_cur
            best_len = len_cur
            max_like = like
```

```python
# plot contour of marginal log likelihood
plt.contourf(sig_mesh, len_mesh, like_cont,100)
plt.colorbar()
plt.xlabel('sig')
plt.ylabel('ls')
plt.title('marginal log likelihood')
plt.show()
```

### Optimize hyperparameters.

```python
# minimize minus log likelihood using Nelder-Mead
x0 = np.array([1,1])
args = (tp,tp_vals)
res = minimize(minus_like_hyp, x0, args, method='nelder-mead', \
               options={'xtol': 1e-8, 'disp': True})

print(res.x)
```
